```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_train_en= pd.read_csv('Dataset/English/en_train.csv')
df_train_es= pd.read_csv('Dataset/Spanish/es_train.csv')
```

```python
# Plot pie chart for binary column
binary_counts = df_train_en['binary'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(binary_counts, labels=binary_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Binary Class Distribution')
plt.show()

# Plot pie chart for multiclass column
multiclass_counts = df_train_en['multiclass'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(multiclass_counts, labels=multiclass_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Multiclass Class Distribution')
plt.show()
```

```python
# Plot pie chart for binary column
binary_counts = df_train_es['binary'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(binary_counts, labels=binary_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Binary Class Distribution')
plt.show()

# Plot pie chart for multiclass column
multiclass_counts = df_train_es['multiclass'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(multiclass_counts, labels=multiclass_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Multiclass Class Distribution')
plt.show()
```
